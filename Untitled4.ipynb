```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

data=pd.read_csv('Iris.csv')
data=data.drop(columns='Id')

X=data.drop(columns='Species')
y=data['Species']

Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2)

model=SVC(kernel='linear')

model.fit(Xtrain,ytrain)

pred=model.predict(Xtrain)

plt.scatter(data['PetalLengthCm'][data['Species']=='Iris-setosa'],data['PetalWidthCm'][data['Species']=='Iris-setosa'])
plt.scatter(data['PetalLengthCm'][data['Species']=='Iris-setosa'],data['PetalWidthCm'][data['Species']=='Iris-versicolor'])
plt.scatter(data['PetalLengthCm'][data['Species']=='Iris-setosa'],data['PetalWidthCm'][data['Species']=='Iris-virginica'])
plt.plot(pred)
```
